## Data Statistics

```python
import pandas as pd
import os
annotation_path = "./annotation"
```

```python
frames = []
for root, dirs, files in os.walk(annotation_path):
    for name in files:
            path = os.path.join(root, name)
            frames.append(pd.read_csv(path))
df = pd.concat(frames, ignore_index = True)
df
```

# Class distribution

```python
df.groupby("label").count().track_id.plot(kind='bar')
```

```python
for root, dirs, files in os.walk(annotation_path):
    for name in files:
        path = os.path.join(root, name)
        pd.read_csv(path).groupby('frame_id').count()
```
